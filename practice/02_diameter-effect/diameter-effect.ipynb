Effect of diameter

1m tube with d10, 20, 50, 100

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d10 = pd.read_csv('d10.imp',header=0)
```

```python
d20 = pd.read_csv('d20.imp',header=0)
```

```python
d50 = pd.read_csv('d50.imp',header=0)
```

```python
d100 = pd.read_csv('d100.imp',header=0)
```

```python
dd = pd.DataFrame()
```

```python
dd['x']=d10['freq']
dd['d10']=d10['imp.mag']
dd['d20']=d20['imp.mag']
dd['d50']=d50['imp.mag']
dd['d100']=d100['imp.mag']
```

```python
dd.plot(x = 'x', y = ['d10','d20','d50','d100'])
```

As diameter gets larger, impedance gets dull due to air friction.
Also it makes some frequency shift at higher range.
