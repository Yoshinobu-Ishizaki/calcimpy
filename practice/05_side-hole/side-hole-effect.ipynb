# comparison of side-hole effect

## simple 1m tube impedance

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# module to do peak search
from scipy.signal import argrelextrema
```

```python
smp = pd.read_csv('simple.imp',header=0)
```

```python
smp.plot('freq','imp.mag')
```

Due to existance of side-hole, impedance is changed lot.

```python
hole = pd.read_csv('side-hole.imp',header=0)
```

```python
hole.plot('freq','imp.mag')
```

Side hole effect is small when it is closed.

```python
cls = pd.read_csv('side-hole-close.imp',header=0)
```

```python
cls.plot('freq','imp.mag')
```

Compare three.

```python
dfm = pd.DataFrame()
```

```python
dfm['freq'] = smp['freq']
dfm['simple'] = smp['imp.mag']
dfm['side-hole-open'] = hole['imp.mag']
dfm['side-hole-close'] = cls['imp.mag']
```

```python
dfm.plot('freq',['simple','side-hole-open','side-hole-close'] )
```

It is worth to point out that 1st large peak of 'side-hole-open' is slightly near to 2nd peak of 'simple/close' state.
So that, this mode's pressure is similar to 2nd mode of 'simple/close' state.
